```python
# Function definition
def anagram_checker(word_a, word_b):
    # Convert both words to lowercase using str.lower()
    word_a = word_a.lower()
    word_b = word_b.lower()

    # Sort the characters in both words
    sorted_word_a = sorted(word_a)
    sorted_word_b = sorted(word_b)

    # Compare the sorted lists to determine if they are anagrams
    return sorted_word_a == sorted_word_b

 # Assertion tests
    assert anagram_checker("listen", "silent") == True, "Test failed: 'listen' and 'silent' should be anagrams"
    assert anagram_checker("Part", "trap") == True, "Test failed: 'Part' and 'trap' should be anagrams"
    assert anagram_checker("hello", "bellow") == False, "Test failed: 'hello' and 'bellow' are not anagrams"
    assert anagram_checker("School master", "The classroom") == True, "Test failed: 'School master' and 'The classroom' should be anagrams"
    assert anagram_checker("", "") == True, "Test failed: Two empty strings are considered anagrams"
    
print("All tests passed!")
  
```
